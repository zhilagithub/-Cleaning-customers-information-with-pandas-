```python
df=df.drop(columns="Not_Useful_Column")
df
```

```python
## working on cleaning the last name column 


df["Last_Name"].str.strip("/..._")
```

```python
df["Last_Name"]=df["Last_Name"].str.strip("/..._")
df
```

```python
# clean phone number column


df["Phone_Number"]=df["Phone_Number"].str.replace("[^a-zA-Z0-9]","")

df["Phone_Number"]
```

```python
#### IMPORTANT CODES FOR PHONE NUMBER COLUMN ####

df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x))

df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3] +"-"+ x[3:6]+ "-" +x[6:10])
df["Phone_Number"]
```

```python
###IMPORTANT CODE FOR CLEAING THE PHONE NUMBER COLUMN####

df["Phone_Number"]=df["Phone_Number"]=df["Phone_Number"].str.replace("Na--","")
df["Phone_Number"]=df["Phone_Number"].str.replace("nan--","")
df["Phone_Number"]
```

```python
##### working on cleaning the address column 
```

```python
df[["Address","state","zip_code"]]=df["Address"].str.split(",",2,expand=True)
df[["Address","state","zip_code"]]
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("Yes","Y")
df["Paying Customer"]=df["Paying Customer"].replace("No","N")

df["Paying Customer"]
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("Yes","Y")
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("No","N")

df["Do_Not_Contact"]
```

```python
### remove N/A from the whole dataset
df=df.replace("N/a","")
df=df.replace("NaN","")
df=df.fillna("")
```

```python
df
```

```python
#Create a list of customers who are eligible for company contact 
# by excluding those marked with 'Y' in the 'Do_Not_Contact' column.




# working on Do_Not_Contact column


for i in df.index:
    if df.loc[i, "Do_Not_Contact"] == 'Y':
        df.drop(i, inplace=True)

df    
```

```python
#Create a list of customers who are eligible for company contact 
#by excluding those that dont have phone number in the "Phone_Numbe" column



for i in df.index:
    if df.loc[i, "Phone_Number"] == '':
        df.drop(i, inplace=True)
df       
```
